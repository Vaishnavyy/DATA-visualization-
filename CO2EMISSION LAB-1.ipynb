```python
#WE ARE USING THIS BECAUSE IT IS A QUANTITATIVE VARIABLES
filter_data=emission[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY',
                      'FUELCONSUMPTION_COMB','CO2EMISSIONS']]
filter_data.head()
```

```python
#here we understand correlation of CO2EMISSION with other variables
plt.scatter(filter_data.ENGINESIZE, filter_data.CO2EMISSIONS, color ='pink')
plt.xlabel("Engine Size")
plt.ylabel("CO2 Emission")

plt.show()

plt.scatter(filter_data.CYLINDERS, filter_data.CO2EMISSIONS, color ='gray')
plt.xlabel("Cylinders")
plt.ylabel("CO2 Emission")

plt.show()

plt.scatter(filter_data.FUELCONSUMPTION_CITY, filter_data.CO2EMISSIONS, color ='orange')
plt.xlabel("Fuel Consumption in the City")
plt.ylabel("CO2 Emission")

plt.show()

plt.scatter(filter_data.FUELCONSUMPTION_HWY, filter_data.CO2EMISSIONS, color ='blue')
plt.xlabel("Fuel Consumption in the Highway")
plt.ylabel("CO2 Emission")

plt.show()
```
